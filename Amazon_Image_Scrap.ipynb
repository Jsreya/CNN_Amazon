```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait as wdt
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as ec
from selenium.common.exceptions import StaleElementReferenceException as ser
import os,requests,pandas as pd
import time
```

```python
#creating directory
def directory(directory_name):
    CurrentPath=os.getcwd()
    path=os.path.join(CurrentPath,directory_name)
    if not os.path.exists(path):
        os.makedirs(path)
    print('Directory {0} has been Created\n\n'.format(directory_name))
```

```python
#scrapping images and downloading and other data
def scrapper(driver,directory,page):
    im= str(input("Enter the X_path locator of image:::\t"))
    img=driver.find_elements_by_xpath(im)
    print("\nExtracted x_paths are\n",img[0:10])
    links=[]
    for image in img:
        source=image.get_attribute('src')
        links.append(source)
    print('\nTotal links found',len(links),'\nDownloadable links are:\n',source)
    import shutil
    import requests
    for index,link in enumerate(links):
        print('\nGetting ', index+1,'of', len(links),'Images from the Page to the Local Machine')
        response=requests.get(link)
    
        with open('{0}/ img{1}_{2}{3}.jpeg'.format(directory,directory,index+1,page), 'wb+') as doc:
            doc.write(response.content)
```

```python
#Image Srcrapping Saree
driver=webdriver.Chrome(executable_path=r'C:\Users\hp\cdriver.exe')
driver.get('https://www.amazon.in/')
wait=wdt(driver,14)
element=wait.until(ec.element_to_be_clickable((By.XPATH,"//*[@id='twotabsearchtextbox']")))
element.click()
element.send_keys('Saree')
driver.find_element_by_class_name('nav-input').click()
time.sleep(7)

#creating directory
directory(directory_name= "saree")

#scrapping and downloading the files
initial=1
final=4
for i in range (initial,final+2):
    try:
        scrapper(driver=driver,directory='saree',page=i)                   
#moving to next page and downloading
        time.sleep(6)
        print("Images from page",i,"has been downloaded.\n Moving to Next Page")
        driver.find_element_by_xpath("//li[@class='a-last']").click()
        
    except ser:
        print("WE ARE FACING UNANTED EXCEPTION.RESTARTING THE SERVICE..\n\n")
        error_page=driver.find_element_by_xpath("//li[@class='a-last']").text
        print('Exceptions Faced at',error_page)
        value=driver.find_element_by_xpath("//li[@class='a-last']")
        link=value.get_attribute('href')
        driver.get(link)
        
        scrapper(driver=driver,directory='saree')
        time.sleep(6)
        print("Images from page",i,"has been downloaded.\n Moving to Next Page")
        driver.find_elements_by_class_name('a-last').click()
```

```python
driver.quit()
```
